# Perfume Dupe Finder in Action

This notebook demonstrates the perfected `DupeFinder` algorithm. It enhances raw note intersection with:
- **Log-scaled Rarity Weights**: Rare notes contribute much more to similarity than common notes like Vanilla or Musk.
- **Cross-Layer Support**: Notes matching Top-to-Top receive a 1.0 multiplier, whereas Top-to-Base cross matches get penalized with a 0.6 scale.
- **Flanker vs Clone Heuristics**: If the candidate is from the same brand, it undergoes a `0.9` flanker penalty. Known clone houses (Lattafa, Armaf, etc.) receive a `1.08` boost.
- **Gender Alignment**: Matching gender target demographics yields a `1.05` boost.
